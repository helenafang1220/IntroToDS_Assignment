### 2.2 Correlation Heatmap<a class="anchor" id="2.2"></a>

```python
#correlation
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")


# Compute the correlation matrix
corr = AvPayGap_df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.zeros_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(8, 6))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(250, 15, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=plt.get_cmap('coolwarm'), vmax=1,center=0,
            square=True,annot=True, linewidths=.5, cbar_kws={"shrink": .5}).set(title='Correlation HeatMap')

f.savefig('heatmap.jpg', format='jpg', dpi=600, bbox_inches='tight', pad_inches=0.1,
        facecolor='auto', edgecolor='auto',
        backend=None,)
```

```python
x = AvPayGap_df['Female_Employee'].values.reshape(-1,1)
y = AvPayGap_df['Average_Pay_Gap'].values.reshape(-1,1)
```

### 2.3 Linear Regression <a class="anchor" id="2.3"></a>

```python
regressor = LinearRegression()  
regressor.fit(x,y)
```

```python
r_sq = regressor.score(x, y)
print('coefficient of determination:', r_sq)

print('intercept:', regressor.intercept_)
print('slope:', regressor.coef_)
```

```python
import statsmodels.api as sm
X2 = sm.add_constant(x)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
y_pred = regressor.predict(x)
#print('predicted response:', y_pred, sep='\n')
plt.scatter(x, y, color="black")
plt.plot(x, y_pred, color="blue", linewidth=2)
plt.xlabel("Number of Female Employees")
plt.ylabel("Average Pay Gap(NTD)")

plt.savefig('regression.jpg', format='jpg', dpi=600, bbox_inches='tight', pad_inches=0.1,
        facecolor='auto', edgecolor='auto',
        backend=None,)
```

#### Prediction of average pay gap when there's x amount of female employees.

```python
#make prediction 
y_pred = regressor.predict([[60000]])
print('predicted response:', y_pred, sep='\n')
```
