### Motivation
The dataset downloaded from Taiwan National Health Insurance (NHI), provided stastistics on number of people who got diarrhea that use health insurance to seek for medical help. The aim of analysis is to see whether it's possible to find interesting patterns of behavior when seeking for medical help, and whether it has to do with their locaiton. 

```python
import random, os, sys, shutil
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np
import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### 1 Loading Data  <a class="anchor" id="1"></a>

```python
#Read in csv
df = pd.read_csv("NHI_Diarrhea_eng.csv")
```

```python
#See columns
df.columns
```

```python
# check if there's any empty cells
df.isna().any()
```

```python
x = df
x.head()
```

### 2 Standardise <a class="anchor" id="2"></a>

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#Standardise
x = StandardScaler().fit_transform(x)
```

```python
#Plot in 2D doesn't appear to have clear groups
plt.figure(figsize=(4, 4))
a = plt.plot(x[:,0],x[:,1],"rx")
```

```python
def plot_data(X):
    plt.plot(X[:, 0], X[:, 1], 'k.', markersize=2)

def plot_centroids(centroids, weights=None, circle_color='r', cross_color='k'):
    if weights is not None:
        centroids = centroids[weights > weights.max() / 10]
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='o', s=35, linewidths=8,
                color=circle_color, zorder=10, alpha=0.9)
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=2, linewidths=12,
                color=cross_color, zorder=11, alpha=1)

def plot_clusters(clusterer, X):
    labels = clusterer.predict(X)
    pca = PCA(n_components=2)
    x_2d = pca.fit_transform(X)
    plt.scatter(x_2d[:, 0], x_2d[:, 1], c=labels, alpha=0.3)
    plot_centroids(clusterer.cluster_centers_)
```

### 3 Find appropriate K mean through Elbow Plot <a class="anchor" id="3"></a>

```python
from sklearn import preprocessing
from matplotlib.ticker import FixedLocator, MaxNLocator

normalized_vectors = preprocessing.normalize(df)

scores = [KMeans(n_clusters=i+2).fit(normalized_vectors).inertia_  
          for i in range(20)]

g = sns.lineplot(np.arange(0, 20), scores, marker='o')
g.set_xticks(range(0,20))
plt.xlabel('Number of clusters')
plt.ylabel("Inertia")
plt.title("Inertia of Cosine k-Means versus number of clusters")

plt.savefig("intertia_cosine_kmeans.jpg", dpi=150)
```

### 4 Cluster and Plot <a class="anchor" id="4"></a>

```python
k = 4 # when k=4 four shows the best clustering result
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(x)
plt.figure(figsize=(8, 8))

#Plot clusters onto PCA reduced plot
plot_clusters(kmeans, x)
plt.savefig("k=5.jpg", dpi=150)
```

### 5 Reduce Dimensions & Cluster Again <a class="anchor" id="5"></a>

The image below shows the process of finding the most approprate KMean value, shilhouette score reaches the highest with k=4.

```python
#How many dimensions to reduce to (before clustering)?
num_dimensions = 2

#Reduce dimensions
pca = PCA(n_components=num_dimensions)
x_less_dimensions = pca.fit_transform(x)

#Fit cluster
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(x_less_dimensions)

#Plot results on 2D plot
plt.figure(figsize=(8, 8))
plot_clusters(kmeans, x_less_dimensions)
plt.savefig("k=5_re.jpg", dpi=150)
```

### 6 Examine Clusters <a class="anchor" id="6"></a>

```python
from sklearn.metrics import silhouette_score
```

```python
print(silhouette_score(x, kmeans.labels_, metric='euclidean')) #This may take a while to run
```

```python
#the score improved a lot after reducing the dimensions
print(silhouette_score(x_less_dimensions, kmeans.labels_, metric='euclidean')) #This may take a while to run 
```

```python
PCA_components = pd.DataFrame(x_less_dimensions)
pred = kmeans.predict(PCA_components.iloc[:,:4])
frame = pd.DataFrame(df)
frame['cluster'] = pred
frame.head()
```

```python
df_c = frame.groupby(['cluster'], as_index=False).mean()
df_c
```

```python
#caculate the proportion of diarrhea patients
df_c['portion_diarrhea']= df_c['Number_of_Visit_Cases']/df_c['Total_Number_of_health_insurance_Visit_Cases']
```

```python
df_c
```

```python
#check the details in the Age column
age_df = pd.DataFrame(frame.groupby(['cluster','Age'])['Age'].count())
age_df
```

```python
loc_df = pd.DataFrame(frame.groupby(['cluster','Location'])['Location'].count())
loc_df
```
